```python

## Getting paginated repos:
## If the project has only one repo, you do not need this segment

import requests

# Replace 'owner_username' with the GitHub username of the owner
owner_username = ''

# GitHub API endpoint for repositories of a specific user
api_url = f'https://api.github.com/users/{owner_username}/repos'

try:
    # List to store all repository names
    all_repository_names = []
    
    # Variable to store the page number
    page = 1
    
    while True:
        # Make a GET request to the GitHub API for the current page
        response = requests.get(api_url, params={'page': page, 'per_page': 100})  # Fetch 100 repositories per page
        
        # Check if the request was successful (HTTP status code 200)
        if response.status_code == 200:
            # Parse the JSON response
            repositories = response.json()
            
            # If there are no more repositories on the current page, break the loop
            if len(repositories) == 0:
                break
            
            # Extract repository names from the current page and add to the list
            repository_names = [repo['name'] for repo in repositories]
            all_repository_names.extend(repository_names)
            
            # Move to the next page
            page += 1
        else:
            # Print an error message if the request was not successful
            print(f'Error: Unable to fetch repositories. Status code: {response.status_code}')
            break
    
    # Print all repository names
    print(f'Repositories of {owner_username}: {all_repository_names}')
except Exception as e:
    # Handle exceptions
    print(f'An error occurred: {e}')
```
